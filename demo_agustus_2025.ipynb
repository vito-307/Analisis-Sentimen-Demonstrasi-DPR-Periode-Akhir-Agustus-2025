# **Preprocessing FullText**

```python
import re
import string
import nltk

#fungsi menghapus url
def remove_url(text):
  if text is not None and isinstance(text, str):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'', text)
  else:
    return text

#fungsi menghapus HTML
def remove_html(text):
  if text is not None and isinstance(text, str):
    html = re.compile(r'<.*?>')
    return html.sub(r'', text)
  else:
    return text

#fungsi hapus emoji
def remove_emoji(text):
  if text is not None and isinstance(text, str):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric
                               u"\U0001F800-\U0001F8FF"  # Supplemental
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  #Symbols and Pyctograph Extended-A
                               u"\U0001F1E0-\U0001F1FF"  #flags
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
  else:
    return text

#fungsi menghapus simbol
def remove_symbol(text):
  if text is not None and isinstance(text, str):
    text = re.sub(r'[^\w\s]', '', text)
    return text

#fungsi hapus username
def remove_username(text):
  if text is not None and isinstance(text, str):
    username = re.compile(r'@[^\s]+')
    return username.sub(r'', text)
  else:
    return text

df_selected['cleaning'] = df_selected['full_text'].apply(lambda x: remove_url(x))
df_selected['cleaning'] = df_selected['cleaning'].apply(lambda x: remove_html(x))
df_selected['cleaning'] = df_selected['cleaning'].apply(lambda x: remove_emoji(x))
df_selected['cleaning'] = df_selected['cleaning'].apply(lambda x: remove_symbol(x))
df_selected['cleaning'] = df_selected['cleaning'].apply(lambda x: remove_username(x))

df_selected
```

# **Case Folding**

```python
def case_folding(text):
  if isinstance(text, str):
    return text.lower()
  else:
    return text

df_selected['case_folding'] = df_selected['cleaning'].apply(case_folding)
df_selected
```

# **Words Normalization**

```python
# Manual word normalization function
def normalize_words_manual(text):
    if isinstance(text, str):
        replacements = {
            # kata sambung / penghubung
            "yg": "yang",
            "dgn": "dengan",
            "sm": "sama",
            "biar": "supaya",
            "utk": "untuk",
            "krn": "karena",
            "dlm": "dalam",
            "dr": "dari",
            "tp": "tapi",
            "jg": "juga",
            "aja": "saja",
            "gmn": "gimana",
            "trs": "terus",
            "kl": "kalau",
            "spy": "supaya",

            # kata kerja & makna umum
            "nggak": "tidak",
            "gak": "tidak",
            "ga": "tidak",
            "tdk": "tidak",
            "udah": "sudah",
            "sdh": "sudah",
            "dah": "sudah",
            "blm": "belum",
            "belom": "belum",
            "bkn": "bukan",
            "hrs": "harus",
            "bs": "bisa",
            "ngapain": "untuk apa",
            "ngapa": "kenapa",

            # konteks demo & politik
            "brimob": "brimob",
            "mako": "mako",
            "dpr": "dpr",
            "demo": "demonstrasi",
            "unjukrasa": "unjuk rasa",
            "aksi": "aksi",
            "mahasiswa": "mahasiswa",
            "polisi": "polisi",
            "rakyat": "rakyat",
            "tni": "tni",
            "gasairmata": "gas air mata",
            "orasi": "orasi"
        }

        # ganti kata tidak baku
        for key, val in replacements.items():
            text = text.replace(key, val)
        return text
    else:
        return text

# Terapkan ke dataframe
df_selected['normalized'] = df_selected['case_folding'].apply(normalize_words_manual)
display(df_selected.head())
```

# **Visualisasi Topik Pembicaraan**

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
from collections import Counter

# Fungsi untuk membuat diagram batang frekuensi n-gram
def plot_ngram_frequencies(texts, ngram_range=(3,3), top_n=10, color='skyblue'):
    vectorizer = CountVectorizer(ngram_range=ngram_range)
    X = vectorizer.fit_transform(texts)

    # Hitung total frekuensi setiap n-gram
    ngram_counts = X.sum(axis=0).A1
    ngram_vocab = vectorizer.get_feature_names_out()
    freq_dist = dict(zip(ngram_vocab, ngram_counts))

    # Ambil n-gram dengan frekuensi tertinggi
    top_ngrams = Counter(freq_dist).most_common(top_n)
    labels, values = zip(*top_ngrams)

    # Visualisasi
    plt.figure(figsize=(10, 6))
    bars = plt.barh(labels[::-1], values[::-1], color=color)  # urutkan dari kecil ke besar
    plt.title(f'Top {top_n} {ngram_range[0]}-gram Frekuensi')
    plt.xlabel('Frekuensi')

    # Tambahkan label di ujung setiap bar
    for bar in bars:
        width = bar.get_width()
        plt.text(width, bar.get_y() + bar.get_height()/2,
                 f'{width}', va='center')

    plt.tight_layout()
    plt.show()

# Visualisasi 3-gram
plot_ngram_frequencies(df_selected['normalized'], ngram_range=(3,3), top_n=10, color='lightcoral')
```

# **Stopword Removal**

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

# Ambil stopwords bawaan nltk (indonesia)
stop_words = set(stopwords.words('indonesian'))

# Tambahkan stopwords custom sesuai konteks (demo, DPR, Brimob)
custom_stopwords = {
    "aja", "banget", "nih", "sih", "dong", "deh", "ya", "nya",
    "kayak", "yg", "dpr", "brimob", "mako", "demo", "polisi", "tni"
}
stop_words = stop_words.union(custom_stopwords)

def remove_stopwords(text):
    return [word for word in text if word not in stop_words]

# Ubah hasil list jadi string
df_selected['stopword removal'] = df_selected['tokenized'].apply(
    lambda x: " ".join(remove_stopwords(x))
)

df_selected.head(5)
```

```python
# Fungsi untuk membuat diagram batang frekuensi n-gram
def plot_ngram_frequencies(texts, ngram_range=(3,3), top_n=10, color='skyblue'):
    vectorizer = CountVectorizer(ngram_range=ngram_range)
    X = vectorizer.fit_transform(texts)

    # Hitung total frekuensi setiap n-gram
    ngram_counts = X.sum(axis=0).A1
    ngram_vocab = vectorizer.get_feature_names_out()
    freq_dist = dict(zip(ngram_vocab, ngram_counts))

    # Ambil n-gram dengan frekuensi tertinggi
    top_ngrams = Counter(freq_dist).most_common(top_n)
    labels, values = zip(*top_ngrams)

    # Visualisasi
    plt.figure(figsize=(10, 6))
    bars = plt.barh(labels[::-1], values[::-1], color=color)  # urutkan dari kecil ke besar
    plt.title(f'Top {ngram_range[0]}-gram Frekuensi (Setelah Stopword Removal)')
    plt.xlabel('Frekuensi')

    # Tambahkan label di ujung setiap bar
    for bar in bars:
        width = bar.get_width()
        plt.text(width, bar.get_y() + bar.get_height()/2,
                 f'{width}', va='center')

    plt.tight_layout()
    plt.show()

# Visualisasi 3-gram
plot_ngram_frequencies(df_selected['stopword removal'], ngram_range=(3,3), top_n=10, color='lightcoral')
```

```python
from collections import Counter

# Group by date and find the most frequent word for each date
def most_frequent_word_by_date(df):
    daily_word_counts = {}
    for date, group in df.groupby('date'):
        all_words_for_date = [word for text in group['stopword removal'] for word in text.split()]
        if all_words_for_date:
            word_counts = Counter(all_words_for_date)
            most_common = word_counts.most_common(1)
            if most_common:
                daily_word_counts[date] = most_common[0][0]
            else:
                daily_word_counts[date] = None # Handle cases with no words after removal
        else:
            daily_word_counts[date] = None # Handle cases with no text for the date
    return daily_word_counts

most_frequent_words = most_frequent_word_by_date(df_selected)

# Display the most frequent word for each date
print("Most frequent word by date:")
for date, word in most_frequent_words.items():
    print(f"{date}: {word}")
```

# **Labelling Lexicon Based**

```python
import pandas as pd
import requests

# Unduh kamus leksikon positif dan negatif dari GitHub
positive_url = "https://raw.githubusercontent.com/fajri91/InSet/master/positive.tsv"
negative_url = "https://raw.githubusercontent.com/fajri91/InSet/master/negative.tsv"

positive_lexicon = set(pd.read_csv(positive_url, sep="\t", header=None)[0])
negative_lexicon = set(pd.read_csv(negative_url, sep="\t", header=None)[0])

def determine_sentiment(text):
    if isinstance(text, str):
        positive_count = sum(1 for word in text.split() if word in positive_lexicon)
        negative_count = sum(1 for word in text.split() if word in negative_lexicon)
        sentiment_score = positive_count - negative_count
        sentiment = "Negatif" if sentiment_score <= 0 else "Positif"
        return sentiment_score, sentiment
    return 0, "Negatif"

df_selected[['Score', 'Sentiment']] = df_selected['stopword removal'].apply(
    lambda x: pd.Series(determine_sentiment(x))
)

df_selected.head(5)
```

```python
import matplotlib.pyplot as plt

# Calculate the frequency of each sentiment
sentiment_counts = df_selected['Sentiment'].value_counts()

# Calculate the percentage of each sentiment
sentiment_percentages = (sentiment_counts / len(df_selected)) * 100

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(sentiment_counts.index, sentiment_counts.values, color=['skyblue', 'lightcoral'])
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.title('Distribusi Sentimen Publik')
plt.xticks(rotation=0)

# Add percentage labels on top of the bars
for bar, percentage in zip(bars, sentiment_percentages.values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 5,
             f'{percentage:.2f}%',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
